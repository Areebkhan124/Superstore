```python
for col in ['Sales', 'Profit']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: {len(outliers)} outliers")
```

```python
import scipy.stats as stats

for col in ['Sales', 'Profit']:
    z_scores = np.abs(stats.zscore(df[col]))
    outliers = df[z_scores > 3]
    print(f"{col}: {len(outliers)} outliers (Z > 3)")
```

```python
outliers = df[(df['Discount'] > 0.8) | (df['Profit'] < 0)]
print(f"Domain-based Outliers: {len(outliers)} rows found")
```

```python
for col in ['Sales', 'Profit']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
for col in ['Sales', 'Profit']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower, lower,
                       np.where(df[col] > upper, upper, df[col]))
```

```python
df_clean = df.copy()
print("All outliers kept for further analysis as genuine cases.")
```

# Impact of outliers

```python
### Effect On Statistical Measures

#  Outliers affect statistical measures by increasing the mean and standard deviation, giving a false impression of data variability.


### Effect On Machine learning Model

# Outliers negatively impact model training, causing biased coefficients and lower predictive performance.


### Effect On Visualization

# Outliers distort visualizations by stretching scales, making normal data trends harder to observe.


```
